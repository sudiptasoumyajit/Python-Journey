```python
#import the required packages 
import pandas as pd
import matplotlib.pyplot as plt

#Read the csv file.
df = pd.read_csv("Titanic.csv")
df
```

# Activity 1 - Plotting histogram showing the passenger of different age groups who survived


```python

#Remove the rows from the data set where there are NaN values
df = df.dropna()

#Find out the passengers from the Titanic who survived in Titanic.
passengers_survived = df.loc[df['Survived'] == 1]
passengers_survived


```

```python
#Plot a histogram showing the passenger of titanic from different age groups who survived.
plt.figure(figsize=(7,7))
plt.hist(passengers_survived["Age"], bins=10)
plt.title("Number of survival as per age group")
plt.ylabel("count")
plt.xlabel("age")
plt.show()
```

Conclusion: So from the Histogram we, can conclude that, most passengers who survived were in the age group of 20-40.

# Activity 3 - Find the total number of passengers who were travelling and total number of passengers who died from class1, class 2, and class 3. 

```python
#predefine code for image
Image(filename='titanic_classes.png') 
#predefine code end
```

```python
survived1=df.loc[(df['Survived']==1)&(df['Pclass']==1)]

total_class1=df.loc[df['Pclass']==1]['Pclass'].count()
total_class1

survive_in_class1=survived1['Pclass'].count()
survive_in_class1
```

```python
survived2=df.loc[(df['Survived']==1)&(df['Pclass']==2)]

total_class2=df.loc[df['Pclass']==2]['Pclass'].count()
total_class2

survive_in_class2=survived2['Pclass'].count()
survive_in_class2
```

```python
survived3=df.loc[(df['Survived']==3)&(df['Pclass']==2)]

total_class3=df.loc[df['Pclass']==3]['Pclass'].count()
total_class3

survive_in_class3=survived3['Pclass'].count()
survive_in_class3
```

```python
total_passengers=[total_class1,total_class2,total_class3]
total_survival=[survive_in_class1,survive_in_class2,survive_in_class3]
index=['total_vs_survived_class1','total_vs_survived_class2','total_vs_survived_class3']
```

```python
plt.bar(index,total_passengers,bottom=total_survival,color='darkblue',label='Total Mans')
plt.bar(index,total_survival,color='darkgreen',label='Total Survived')
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```
