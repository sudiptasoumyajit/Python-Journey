```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as chart

dataframe=pd.read_csv('db.csv')

getTable = dataframe.head(5)
print(getTable)

chart.xlabel('Org name')
chart.xticks(rotation='vertical')
chart.ylabel('No of satellites')

chart.bar(getTable['Country/Organization Name'],getTable['Satellites In Orbit'],width=0.5,color=('green','red','yellow','black','pink'))
```
